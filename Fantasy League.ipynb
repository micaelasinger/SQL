```python
c.execute('''
    CREATE TABLE IF NOT EXISTS matches 
    ([team] TEXT KEY, 
        [opponent_team] TEXT, 
        [game_week] INTEGER,
        [match_number] INTEGER KEY,
        [date_of_play] DATE, 
        [weekday_of_play] TEXT, 
        [hour_of_play] INTEGER, 
        [result] TEXT, 
        [goals_scored] INTEGER, 
        [goals_conceded] INTEGER, 
        [total_goals] INTEGER, 
        [location] TEXT, 
        [referee] TEXT, 
        [crowd] INTEGER
    )
''')
```

```python
c.execute('''
    CREATE TABLE IF NOT EXISTS stats 
    ([match_number] INTEGER PRIMARY KEY, 
        [yellow_cards] INTEGER, 
        [red_cards] INTEGER,
        [offsides] INTEGER,
        [corners] INTEGER
    )
''')
```

```python
c.execute('''
    CREATE TABLE IF NOT EXISTS teams
          ([name] TEXT PRIMARY KEY,
           [budget] INTEGER,
           [home] TEXT
           )
''');
```

```python
c.execute('''
    CREATE TABLE IF NOT EXISTS scorers
          ([id] INTEGER,
           [name] TEXT,
           [goal_type] TEXT,
           [goal_minute] INTEGER,
           [match_number] INTEGER KEY
           )
''');
```

## Data Analysis

```python
##Count the total scored goals by nationality
c.execute('''
          SELECT dp.nationality, 
                 count(*) AS total_goals
          FROM scorers s 
          INNER JOIN dim_players dp
          ON s.id=dp.id
          GROUP BY nationality
          ORDER BY 2 DESC
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```

```python
##Get the type of referee according to the yellow cards that he provided
c.execute('''
          SELECT DISTINCT m.referee, s.yellow_cards,
                 CASE WHEN yellow_cards >= 2 THEN 'Hard Referee' else 'Fair Referee' END AS 'Referee_Type' 
          FROM matches m 
          INNER JOIN stats s
          ON m.match_number = s.match_number
          ORDER BY 2 DESC
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```

```python
##Select all the goals made by players whose first letter of their name starts with R
c.execute('''
          SELECT *
          FROM scorers s
          WHERE s.name LIKE 'R%'
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```

```python
##Count the number of goals conceded per goalkeeper, his saves and his fantasy points
c.execute('''
          SELECT g.goalkeeper_name, sum(m.goals_conceded), g.goalkeeper_saves, g.goalkeeper_fantasy_points
          FROM goalkeepers g
          INNER JOIN dim_players dp
          ON g.id = dp.id
          INNER JOIN matches m
          ON g.match_number=m.match_number
          AND dp.team = m.team
          GROUP BY g.goalkeeper_name, g.goalkeeper_saves, g.goalkeeper_fantasy_points
          ORDER BY 2 DESC
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```

```python
#Show each team's budget, crowd and city
c.execute('''
          SELECT t.name, t.home, t.budget, m.crowd
          FROM teams t
          INNER JOIN matches m
          ON t.name=m.team
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```

```python
#Sum the goals scored away and home
c.execute('''
          SELECT m.location,
                 SUM(goals_scored)
          FROM matches m
          GROUP BY m.location
          ''')
df = pd.DataFrame(c.fetchall())
df.head()
```
